```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
car_data = pd.read_csv('mtcars.csv')
car_data
```

```python
car_data.info()
```

```python
car_data.describe()
```

```python
car_data.shape
```

```python
car_data['mpg'].plot(kind = 'hist',bins = 20,title='Efficiency distribution')
```

```python
max_eff = car_data['mpg'].max()
print('cars with highest fuel efficiency:\n',car_data.loc[car_data['mpg'] == max_eff])
```

```python
car_data['am'].nunique()
```

```python
car_data['am'].value_counts()
```

```python
print('Average efficiency for automatic: ',np.mean(car_data['mpg'][car_data['am'] ==1]))
print('Average efficiency for manual: ',np.mean(car_data['mpg'][car_data['am'] ==0]))
```

```python
plt.scatter(car_data['mpg'],car_data['wt'])
plt.title('mpg vs weight')
plt.show()
```

```python
worst_hp = car_data['hp'].min()
print('cars with worst horse power: ', car_data[car_data['hp'] == worst_hp])
```
