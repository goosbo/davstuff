```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('datasets/sentiment.csv')
data
```

```python
#1. In the sentiment feature and text, trim white spaces and convert to lowercase.
data['Sentiment'] = data['Sentiment'].str.strip().str.lower()
data['Text'] = data['Text'].str.strip().str.lower()
data
```

```python
# 2. What are the sentiment categories? Tabulate and visualize the number of texts for each category.
print(f"Sentiment categories: \n")
print(data['Sentiment'].unique())
texts = data.groupby('Sentiment')['Text'].count().sort_values(ascending = False).head(10)
print(texts)
plt.figure(figsize=(12,6))
sns.barplot(x = texts.index, y= texts.values, palette = 'viridis')
plt.title('Number of texts in each Category')
plt.xlabel('Categories', fontsize = 10)
plt.ylabel('Number of texts', fontsize = 10)
plt.xticks(fontsize=10)  
plt.yticks(fontsize=10)
plt.show()
```

Positive is the most common sentiment, followed by joy, excitement contentment and gratitude.

```python
# 3. Tabulate and Visualize the text length for each sentiment category
data['txtlen'] = data['Text'].str.len()
sentiment_txt = data.groupby('Sentiment')['txtlen'].mean().sort_values(ascending = False).head(10)
sentiment_txt
```

```python
sns.barplot(y=sentiment_txt.index,x=sentiment_txt,orient='h')
plt.title('Sentiments sorted by text length')
plt.show()
```

winter magic is the sentiment with the highest average text length followed by whispers of the past and thrilling journey.

```python
# 4. Find and display the text(s) with the maximum number of likes.
print('Texts with most likes:')
most_like = data['Likes'].max()
print('Most likes: ',most_like)
mostlike_text = data['Text'][data['Likes'] == most_like]
for t in mostlike_text:
    print(t)
```

```python
# 5. Find and display the text(s) with the maximum reposts. what is its sentiment?
print('Texts with most reposts:')
most_repost = data['Retweets'].max()
print('Most reposts: ',most_repost)
mostrepost_text = data[['Text','Sentiment']][data['Likes'] == most_repost]
mostrepost_text
```

```python
# 6. Tabulate and Visualize the top 5 countries that have posted the most texts with a positive sentiment.
positive_txts = data[data['Sentiment'] == 'positive']
contr_positive = positive_txts['Country'].value_counts()[:5]
print('Country with most positive sentiment texts')
contr_positive
```

```python
sns.barplot(x=contr_positive.index,y=contr_positive)
plt.title('Countrys with most positive texts')
plt.show()
```

```python
# Find and display all the text(s) with the word â€˜Travelâ€™ in their hashtag.
print('Texts with \'Travel\' in their hashtags')
travel = data['Text'][data['Hashtags'].str.contains('Travel')]
for t in travel:
    print(t)
```

```python
# 8. Find and display the most popular (most number of likes) posts (top 5) from the year 2020 which were of a negative sentiment
data[(data['Year'] == 2020) &(data['Sentiment'] == 'negative')]
```

There is no comment in 2020 with a negative label

```python
# 9. Find and display a list of Users and their posts whose texts are labeled â€˜Happinessâ€™ sentiment
print('List of Users and Texts with sentiments labelled happiness')
data[['User','Text']][data['Sentiment'] == 'happiness']
```

```python
#10. Tabulate and visualize the years and countrywide tweets of the sentiments. Pick tweets with sentiments, â€˜Positive,â€™ â€˜Negative,â€™ and â€˜Neutralâ€™ only.
year_country_data = data[['Year','Country','Sentiment']][(data['Sentiment'] == 'positive') |(data['Sentiment'] == 'negative') |(data['Sentiment'] == 'neutral') ]
print(year_country_data['Year'].unique())# only one year has positive, negative or neutral
year_country_data['Country'] = year_country_data['Country'].str.strip()

sns.barplot(x=year_country_data['Sentiment'].value_counts().index,y=year_country_data['Sentiment'].value_counts().values)
plt.title('Sentiment Distribution in 2023')
plt.show()
```

In the year 2023 the most common sentiment was positive.

```python
print('Sentiment distribution for 2023')
year_country_data['Sentiment'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
country_data = year_country_data.drop('Year',axis=1).groupby('Country')['Sentiment'].value_counts().plot(kind='bar',ax=ax)
plt.title('Country wise Sentiment Distribution')
plt.ylabel('Count')
plt.xlabel('Country')
plt.show()
```

USA has the most positive posts by country followed by UK and Canada. The distribution can be skewed by the bias in number of texts per country
