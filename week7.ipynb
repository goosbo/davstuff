```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.color_palette("Set2")
ipl_data = pd.read_csv('datasets/matches.csv')
ipl_data
```

```python
# 1. Find the names of teams that are part of IPL.
print('Teams participating in the IPL are:')
print(ipl_data['team1'].unique())
```

```python
#2. Visualize the total number of matches won by each team. 
plt.pie(ipl_data['winner'].value_counts(),labels=ipl_data['winner'].value_counts().index, colors=sns.color_palette('Set2'),autopct='%1.0f%%')
plt.title('Total Number of Matches one by each team')
plt.show()
```

```python
sns.barplot(y=ipl_data['winner'].value_counts().index,x=ipl_data['winner'].value_counts(),orient='h')
```

```python
# 3. List the names of umpires who have served as second umpires and the number of matches.
print('List of second umpires and their respective no. of matches')
ipl_data['umpire2'].value_counts()
```

```python
# 4. Visualize how many matches were held in each venue. 
sns.barplot(y=ipl_data['venue'].value_counts().index,x=ipl_data['venue'].value_counts(),orient='h')
plt.title('Number of matches in each venue')
plt.xlabel('Number of matches')
plt.ylabel('Venues')
plt.show()
```

```python
# 5. Visualize the top 10 players identified as the player of the match for the most significant number of matches.
top_players = ipl_data['player_of_match'].value_counts()[:10]
sns.barplot(x=top_players.values,y=top_players.index,orient='h')
plt.title('Top 10 players vs number of player of the matches')
plt.xlabel('top 10 matches')
plt.ylabel('no. of potms')
plt.show()
```

```python
# 6. Visualize the bottom 10 venues where the least number of matches were played. 
worst_venues = ipl_data['venue'].value_counts()[::-1][:10][::-1]
sns.barplot(x=worst_venues.values,y=worst_venues.index,orient='h')
plt.title('Worst 10 Venues')
plt.xlabel('Venues')
plt.ylabel('no. of matches held')
plt.show()
```

```python
# 7. Visualize the number of wins of each team for the venue of â€˜MA Chidambaram Stadium, Chepaukâ€™
stadium_winners = ipl_data['winner'][ipl_data['venue']== 'MA Chidambaram Stadium, Chepauk'].value_counts()
plt.pie(stadium_winners,labels=stadium_winners.index, colors=sns.color_palette('Set2'),autopct='%1.0f%%')
plt.title('Total Number of Matches one by each team in MA Chidambaram Stadium, Chepauk')
plt.show()
```

```python
sns.barplot(x=stadium_winners.values,y=stadium_winners.index,orient='h')
plt.title('Winners in MA Chidambaram Stadium, Chepauk')
plt.xlabel('Number of wins')
plt.ylabel('teams')
plt.show()
```

```python
# 8. Find and list the venues where the match resulted in a tie.
print('Venues were matches resulted in a tie:')
print(ipl_data['venue'][ipl_data['result']=='tie'].values)
```

```python
# 9. List and visualize the best teams(top 5), season-wise.  
seasons = ipl_data.groupby('season')
for i in seasons:
    print(i['winner'].value_counts()[:5])
```

```python
# 10. Analyse and visualize the relationship between winning the toss and winning the match.
win_toss_wincount = len(ipl_data[ipl_data['toss_winner']==ipl_data['winner']])
sns.barplot(y=[win_toss_wincount,len(ipl_data)-win_toss_wincount],x=['won match','lost match'])
plt.title('Match result for teams that won the toss')
plt.ylabel('No. of matches')
plt.xlabel('Match result')
plt.show()
# Winning the toss gives the team a slight advantage in the match but the advantage is minimal
```
