```python
# 1. In the sentiment feature and text, trim white spaces and convert to lowercase.
df['Text'] = df['Text'].str.strip().str.lower()
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()
df['Country'] = df['Country'].str.strip()
```

```python
print('New dataframe: ')
df
```

```python
# 2. What are the sentiment categories? Tabulate and visualize the number of texts for each category.
print(f"Sentiment categories: \n")
print(df['Sentiment'].unique())
texts = df.groupby('Sentiment')['Text'].count().sort_values(ascending = False).head(10)
print(texts)
plt.figure(figsize=(12,6))
sns.barplot(x = texts.index, y= texts.values, palette = 'viridis')
plt.title('Number of texts in each Category')
plt.xlabel('Categories', fontsize = 10)
plt.ylabel('Number of texts', fontsize = 10)
plt.xticks(fontsize=10)  
plt.yticks(fontsize=10)
plt.show()
```

# Observation:
1. Most Frequent Categories:

"Positive" and "Joy" have the highest number of texts, both around 45.

"Excitement" follows closely with slightly fewer texts.

2. Moderate Categories:

"Contentment," "Gratitude," and "Neutral" fall in the mid-range, each having roughly half the number of texts as the top categories.

2. Less Frequent Categories:

"Curiosity," "Serenity," "Happy," and "Nostalgia" appear least frequently, with "Nostalgia" having the fewest entries.

```python
# 3. Tabulate and Visualize the text length for each sentiment category.
df['Text_length'] = df['Text'].apply(len)
text_lengths = df.groupby('Sentiment')['Text_length'].mean().sort_values(ascending = False).head(10)
plt.figure(figsize = (12,6))
sns.barplot(x=text_lengths.index, y= text_lengths.values, palette = 'magma')
plt.xlabel('Category')
plt.ylabel('Mean text length for each category')
plt.xticks(rotation = 45)
plt.show()
```

The X-axis represents different text categories.

The Y-axis represents mean text length for each category.

The bar height shows the average length of text in each category.

# Observations:
1. Winter magic has the highest mean text length.

2. Mesmerizing has the lowest mean text length.

3. Other categories fall in between in descending order.                                    

```python
# 4. Find and display the text(s) with the maximum number of likes.
likes = df.groupby('Text')['Likes'].sum().sort_values(ascending = False).head(10)
plt.figure(figsize = (18,16))
sns.barplot(y= likes.index, x = likes.values, palette = 'Blues')
plt.xticks(fontsize = 35)
plt.yticks(fontsize = 25)
plt.xlabel('Number of likes', fontsize = 30)
plt.ylabel('Texts with most number of likes', fontsize = 30)
plt.title('Texts with maximum number of likes', fontsize = 40)
plt.show()
```

The Y-axis (vertical) represents different text entries (phrases or sentences).

The X-axis (horizontal) represents the number of likes each text received.

The bar length indicates the number of likes a particular text received.

# Observations:
1. The topmost text ("a playful escapade...") received the highest number of likes, around 140.

2. The bottom-most text ("embraced by the hopeful dawn...") received the least number of likes among the top-ranked texts, around 80.

3. The texts are sorted in descending order of likes, meaning the most liked texts appear at the top, and the least liked among them appears at the bottom.

```python
# 5. Find and display the text(s) with the maximum reposts. what is its sentiment?
posts = df.groupby('Text')['Retweets'].sum().sort_values(ascending = False).head(10)
plt.figure(figsize = (18,16))
sns.barplot(y= posts.index, x = posts.values, palette = 'Reds')
plt.xticks(fontsize = 35)
plt.yticks(fontsize = 25)
plt.xlabel('Number of Reposts', fontsize = 30)
plt.ylabel('Texts with most number of Reposts', fontsize = 30)
plt.title('Texts with maximum number of Reposts', fontsize = 40)
plt.show()

```

X-axis (Horizontal): "Number of Reposts" â€“ Represents the count of reposts for each text.

Y-axis (Vertical): "Texts with most number of Reposts" â€“ Displays different text entries that have been reposted the most.

Color Scheme: The bars are shaded in different intensities of red (palette='Reds'), where darker shades indicate a higher number of reposts.

# Observations:

1. The top text has the highest number of reposts, exceeding 70.

2. The bottom-most text has the least reposts among the displayed entries but still a significant amount.

3. The bars decrease in size as you move from bottom to top, indicating a ranking order from most to least reposted.

```python
# 6. Tabulate and Visualize the top 5 countries that have posted the most texts with a positive sentiment.
countries = df[df['Sentiment']=='positive'].groupby('Country')['Sentiment'].count().sort_values(ascending = False).head(5)
print("Top 5 countries that have posted the most texts with a positive sentiment", countries)
plt.figure(figsize = (12,6))
sns.barplot(x = countries.index, y = countries.values, palette = 'Oranges')
plt.xlabel('Countries')
plt.ylabel('Number of texts')
plt.title('Top 5 countries that have posted the most texts with a positive sentiment')
plt.show()

```

Title: "Top 5 countries that have posted the most texts with a positive sentiment"

The chart visualizes the number of texts with a positive sentiment from different countries.

X-axis (Horizontal): "Countries"

Displays the names of the top five countries contributing positive texts.

The listed countries are India, USA (appears twice), UK, and Canada.

Y-axis (Vertical): "Number of texts"

Represents the count of positive sentiment texts posted by each country.

The highest value is 5 texts.

# Observations:

1. USA leads significantly â€“ The USA has the highest number of texts with a positive sentiment (~18), nearly double that of the second-highest country, the UK (~10).

2. Gradual decline â€“ The number of texts decreases progressively from the UK (10) â†’ Canada (8) â†’ India (7) â†’ Australia (2).

3. Australia has the lowest count â€“ Despite being in the top 5, Australia has a much lower number of positive sentiment texts compared to the USA.

4. Balanced middle group â€“ The UK, Canada, and India show a relatively closer range of values (between 7 to 10 texts), indicating some consistency in their contributions.
