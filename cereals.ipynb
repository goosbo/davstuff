```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cereals_data = pd.read_csv('datasets/Cereals.csv')
cereals_data
```

```python
#1. How many cereals(unique brands) are there in the dataset? 
print('No. of cereals: ',cereals_data.shape[0])
```

```python
#2. How many cereals are listed per manufacturer of the cereal? 
print('Cereals per Manufacturer')
cereals_data['mfr'].value_counts()
```

```python
#3. How many hot vs cold cereals? 
print('No. of hot cereals: ',len(cereals_data[cereals_data['type'] == 'H']))
print('No. of cold cereals: ',len(cereals_data[cereals_data['type'] == 'C']))
```

```python
#4. Which is the best and worst cereal? 
print('best cereal')
cereals_data.iloc[cereals_data['rating'].idxmax()]
```

```python
print('worst cereal')
cereals_data.iloc[cereals_data['rating'].idxmin()]
```

```python
# 5. Compare the rating for hot vs cold cereals. Which do people prefer? 
print('Average rating for cold cereal: ',np.mean(cereals_data['rating'][cereals_data['type'] == 'C']))
print('Average rating for hot cereal: ',np.mean(cereals_data['rating'][cereals_data['type'] == 'H']))
fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].boxplot(cereals_data['rating'][cereals_data['type'] == 'C'])
ax[0].set_title('Cold cereal rating')
ax[1].boxplot(cereals_data['rating'][cereals_data['type'] == 'H'])
ax[1].set_title('Hot Cereal rating')
```

```python
#6. Which cereals have the highest fiber and lowest sugar? 
print('cereals with the highest fiber and lower sugar')
cereals_data.sort_values(by =['fiber','sugars'],ascending=[False,True]).iloc[:5]
```

```python
#7. Which Cereals that have more than 3 grams of protein? 
cereals_data[cereals_data['protein']>3]
```

```python
#10. What is the average calories in cereals per manufacturer? 
manufacturers = cereals_data['mfr'].unique()
print(manufacturers)
for m in manufacturers:
    print('Average calories for Manufacturer ',m,': ',np.mean(cereals_data['calories'][cereals_data['mfr']==m]))
```

```python
#11. What is the average nutritional content (calories, sugars, protein, fiber, etc.) across all cereals? 
print('Average calories: ',np.mean(cereals_data['calories']))
print('Average protein: ',np.mean(cereals_data['protein']))
print('Average fat: ',np.mean(cereals_data['fat']))
print('Average sodium: ',np.mean(cereals_data['sodium']))
print('Average fiber: ',np.mean(cereals_data['fiber']))
print('Average carbohydrates: ',np.mean(cereals_data['carbo']))
print('Average sugar: ',np.mean(cereals_data['sugars']))
print('Average potassium: ',np.mean(cereals_data['potass']))
print('Average vitamins: ',np.mean(cereals_data['vitamins']))
```
