```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cereals_data = pd.read_csv('datasets/Cereals.csv')
cereals_data
```

```python
#1. How many cereals(unique brands) are there in the dataset? 
print('No. of cereals: ',cereals_data.shape[0])
```

```python
#2. How many cereals are listed per manufacturer of the cereal? 
print('Cereals per Manufacturer')
cereals_data['mfr'].value_counts()
```

```python
#3. How many hot vs cold cereals? 
print('No. of hot cereals: ',len(cereals_data[cereals_data['type'] == 'H']))
print('No. of cold cereals: ',len(cereals_data[cereals_data['type'] == 'C']))
```

```python
#4. Which is the best and worst cereal? 
print('best cereal')
cereals_data.iloc[cereals_data['rating'].idxmax()]
```

```python
print('worst cereal')
cereals_data.iloc[cereals_data['rating'].idxmin()]
```

```python
# 5. Compare the rating for hot vs cold cereals. Which do people prefer? 
print('Average rating for cold cereal: ',np.mean(cereals_data['rating'][cereals_data['type'] == 'C']))
print('Average rating for hot cereal: ',np.mean(cereals_data['rating'][cereals_data['type'] == 'H']))
fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].boxplot(cereals_data['rating'][cereals_data['type'] == 'C'])
ax[0].set_title('Cold cereal rating')
ax[1].boxplot(cereals_data['rating'][cereals_data['type'] == 'H'])
ax[1].set_title('Hot Cereal rating')
```

```python
#6. Which cereals have the highest fiber and lowest sugar? 
print('cereals with the highest fiber and lower sugar')
cereals_data.sort_values(by =['fiber','sugars'],ascending=[False,True]).iloc[:5]
```

```python
#7. Which Cereals that have more than 3 grams of protein? 
cereals_data[cereals_data['protein']>3]
```

```python
#8. Tabulate and visualize Cereals by display Shelf.
color = ["red","yellow","blue"]
plt.bar(cereals_data['shelf'].value_counts().index,cereals_data['shelf'].value_counts(),color = color)
plt.title('Cereals by Display Shelf')
plt.xlabel('Shelf')
plt.ylabel('Number of Cereals')
plt.show()
cereals_data['shelf'].value_counts()
```

```python
#9.How does the sugar content vary across different brands?
pd.to_numeric(cereals_data['sugars'][cereals_data['sugars']!='?']).plot(kind='box',title='Variation of Sugar across all brands')
# varies from 3 grams to 11 grams with a mean of 7
```

```python
#10. What is the average calories in cereals per manufacturer? 
manufacturers = cereals_data['mfr'].unique()
print(manufacturers)
for m in manufacturers:
    print('Average calories for Manufacturer ',m,': ',np.mean(cereals_data['calories'][cereals_data['mfr']==m]))
```

```python
#11. What is the average nutritional content (calories, sugars, protein, fiber, etc.) across all cereals? 
print('Average calories: ',np.mean(cereals_data['calories']))
print('Average protein: ',np.mean(cereals_data['protein']))
print('Average fat: ',np.mean(cereals_data['fat']))
print('Average sodium: ',np.mean(cereals_data['sodium']))
print('Average fiber: ',np.mean(cereals_data['fiber']))
print('Average carbohydrates: ',np.mean(np.float64(cereals_data['carbo'][cereals_data['carbo']!='?'])))
print('Average sugar: ',np.mean(np.float64(cereals_data['sugars'][cereals_data['sugars']!='?'])))
print('Average potassium: ',np.mean(np.float64(cereals_data['potass'][cereals_data['potass']!='?'])))
print('Average vitamins: ',np.mean(cereals_data['vitamins']))
```

```python
#12.What is the relationship between sugar and calories?
invalid_index =np.where(cereals_data['sugars'] == '?')[0]

plt.scatter(np.float64(cereals_data['sugars'].drop(index=invalid_index)),cereals_data['calories'].drop(index=invalid_index))
plt.title('Sugars vs Calories')
plt.show()
# There is a general trend of increase in calories with increase in sugar but the correlation is not high
```

```python
#13.How does the sugar content compare in cereals with high vs. low ratings?
invalid_index =np.where(cereals_data['sugars'] == '?')[0]

plt.scatter(np.float64(cereals_data['sugars'].drop(index=invalid_index)),cereals_data['rating'].drop(index=invalid_index))
plt.title('Rating vs Calories')
plt.show()
# higher the rating of the cereals, lower is the calories
```

```python
#14.Do cereals marketed as "healthy" (e.g., high fiber, low sugar) tend to have higher ratings?

print('Average rating for healthy cereals: ',np.mean(cereals_data.sort_values(by =['fiber','sugars'],ascending=[False,True]).iloc[:5]['rating']))
print('Total Average Rating: ',np.mean(cereals_data['rating']))
# the top 5 cereals have a higher average rating than the total average
print('cereals with the highest fiber and lower sugar') 
cereals_data.sort_values(by =['fiber','sugars'],ascending=[False,True]).iloc[:5]
```

```python
# 15.Is there a relationship between rating and display shelf?
rating_by_shelf = cereals_data.groupby('shelf')['rating'].mean()
plt.bar(rating_by_shelf.index, rating_by_shelf.values)
plt.title('Average Rating by Display Shelf')
plt.xlabel('Shelf')
plt.ylabel('Average Rating')
plt.show()
# being on shelf 1 gives you the highest tendency to have a good rating, shelf 3 is second and shelf 2 is the lowest
```
