```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
deliveries = pd.read_csv('datasets/deliveries.csv')
deliveries.head()
```

```python
deliveries['batting_team'].unique()
```

```python
#1. Find and visualize the total runs scored by each team
```

```python
team_runs = deliveries.groupby('batting_team')['total_runs'].sum()
```

```python
sns.barplot(y=team_runs.index,x=team_runs,orient='h')
plt.title('Total runs scored by each team')
plt.show()
```

```python
plt.pie(team_runs,labels=team_runs.index,autopct='%.0f%%')
plt.show()
```

```python
#2. Identify and display the names of the top 10 highest-scoring batters with their total runs scored.
batter_total = deliveries.groupby('batter')['batsman_runs'].sum()
batter_total = batter_total.sort_values(ascending=False)
sns.barplot(x=batter_total[:10],y=batter_total.index[:10],orient='h')
plt.title('Top 10 highest scoring batters')
plt.show()
```

```python
# 3. Find and visualize the top 10 bowlers with the most wickets. 
total_wickets = deliveries.groupby('bowler')['is_wicket'].sum()
total_wickets = total_wickets.sort_values(ascending=False)
sns.barplot(x=total_wickets[:10],y=total_wickets.index[:10],orient='h')
plt.title('Top 10 bowlers with most wickets')
plt.show()
```

```python
# 4. Tabulate and visualize the total number of extras conceded by each team.
extra = deliveries.groupby('bowling_team')['extra_runs'].sum()
sns.barplot(x=extra,y = extra.index,orient='h')
plt.title('Total number of extras conceded by each team')
plt.show()
```

```python
# 5. Tabulate and visualize the batsmen (top 5) involved in the most run-outs. 
run_out_rows = deliveries[deliveries['dismissal_kind'] == 'run out']
runout_batter = run_out_rows.groupby('batter').size().sort_values(ascending=False)[:5]
sns.barplot(x=runout_batter,y = runout_batter.index,orient='h')
plt.title('Batsmen involved in most run outs')
plt.show()
```

```python
#6. Tabulate and visualize the top 5 bowlers who conceded the most extras. 
bowler_extra = deliveries.groupby('bowler')['extra_runs'].sum().sort_values(ascending=False)[:5]
sns.barplot(x = bowler_extra,y= bowler_extra.index,orient='h')
plt.title('Bowlers who conceded the most extras')
plt.show()
```

```python
# 7.  Tabulate and visualize the most common dismissal type. 
sns.barplot(x=deliveries['dismissal_kind'].value_counts(),y=deliveries['dismissal_kind'].value_counts().index,orient='h')
plt.title('Dismissal type distribution')
plt.show()
```

```python
#8. Find and display which of the top 3 overs has the highest run rate. 
over_runrate=deliveries.groupby(['match_id','inning','over'])['total_runs'].sum().sort_values(ascending=False)
print('Overs with highest run rates')
over_runrate[:3]
```

```python
#9. Analyze and visualize powerplay performance (first 6 overs). 
powerplay_data = deliveries[deliveries['over'] < 6]
pplay_over=powerplay_data.groupby('over')
pplayruns =pplay_over['total_runs'].mean()
print(pplayruns)
plt.pie(pplayruns,labels=pplayruns.index,autopct='%.0f%%')
plt.title('Mean runs in each over in powerplay')
plt.show()
```

```python
pplaywickets = pplay_over['is_wicket'].mean()
print(pplaywickets)
plt.pie(pplaywickets,labels=pplaywickets.index,autopct='%.0f%%')
plt.title('Mean wickets in each over in powerplay')
plt.show()
```
