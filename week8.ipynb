```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_data = pd.read_csv('datasets/mtcars.csv')
pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',None)

car_data
```

```python
tab1 = pd.crosstab(car_data['cyl'],car_data['gear'])
print('Table summarising number of cars (Cylinders vs Gears)')
tab1
```

```python
sns.heatmap(tab1,annot=True,fmt='d',cmap='YlGnBu')
plt.title('Crosstab Heatmap: Cylinders vs Gears')
plt.xlabel('Gears')
plt.ylabel('Cylinders')
plt.gca().collections[0].colorbar.set_label('Frequency')
```

```python
tab2 = car_data[['model','mpg']].sort_values('mpg',ascending=False)
print('Table displaying Car Models(best to worst fuel efficiency)')
tab2
```

```python
tab3 = car_data.groupby('cyl')['mpg'].mean()
print('Table displaying number of cylinders and average mpg')
tab3
```

```python
sns.barplot(x='cyl',y='mpg',data = car_data,errorbar=None)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='hp',y='mpg',data=car_data)
plt.title('MPG vs HP')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.histplot(car_data['mpg'],bins=10,kde=True)
plt.title('Distribution of MPG')
plt.show()
```

```python
fig,axes = plt.subplots(1,2,figsize=(14,6))
sns.histplot(car_data['mpg'],bins=20,kde=True,ax=axes[0],color='blue')
axes[0].set_title('MPG Distribution')
axes[0].set_xlabel('Miles Per Gallon(MPG)')
axes[0].set_ylabel('Count')

sns.boxplot(x='cyl',y='mpg',data=car_data,ax=axes[1],palette='viridis',)
axes[1].set_title('MPG by Number of Cylinders')
axes[1].set_xlabel('Number of Cylinders')
axes[1].set_ylabel('Miles per Gallon(MPG)')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(car_data[['mpg','hp','wt','qsec']])
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(x='cyl',y='mpg',data=car_data)
plt.title('MPG Distribution by Cylinder Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.regplot(x='wt',y='mpg',data=car_data)
plt.title('MPG vs Weight (Regression)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.swarmplot(x='cyl',y='mpg',data=car_data)
plt.title('MPG by Cylinder Count (Swarm Plot)')
plt.show()
```
