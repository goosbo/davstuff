```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars_data = pd.read_csv('datasets/mtcars.csv')
cars_data
```

```python
cars_data['mpg'].plot()
```

```python
cars_data = cars_data[['cyl','wt','mpg']]
cars_data.plot()
```

```python
cars_data['cyl'].value_counts().plot(kind='bar',color='green')
plt.title('Count of Cars by Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
plt.style.use('ggplot')
```

```python
cars_data.plot(
    kind="scatter",
    x="wt",
    y="mpg",
    figsize=(9,9),
    color="black",
);
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='cyl',y='mpg',data=cars_data)
plt.title('MPG by Cylinder Category')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
plt.hist(cars_data['mpg'])
plt.title("Distribution of MPG in Cars Dataset")
plt.ylabel('Frequency')
plt.xlabel('MPG')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    cars_data['cyl'].value_counts(),
    labels= cars_data['cyl'].value_counts().index,
     autopct='%1.1f%%',
    colors=['skyblue', 'lightgreen', 'orange'],
    startangle=140,
    explode=[0.1, 0, 0]
)
plt.title('Proportion of Cars by Number of Cylinders')
plt.show()
```

```python
cars_data = pd.read_csv('datasets/mtcars.csv')
am_counts = cars_data['am'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    am_counts,
    labels=['Automatic', 'Manual'],  # Custom labels
    autopct='%1.1f%%',
    colors=['gold', 'lightcoral'],
    startangle=90
)
plt.title('Proportion of Cars by Transmission Type')
plt.show()
```

```python
outer_labels =cars_data['cyl'].value_counts().index
outer_sizes = cars_data['cyl'].value_counts().values

inner_labels = cars_data['gear'].value_counts().index
inner_sizes = cars_data['gear'].value_counts().values

# Create nested pie chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(outer_sizes, labels=outer_labels, radius=1, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange'])
ax.pie(inner_sizes, labels=inner_labels, radius=0.7, autopct='%1.1f%%', colors=['pink', 'purple', 'cyan'])
ax.set(aspect="equal")
plt.title('Nested Pie Chart: Cylinders (Outer) and Gears (Inner)')
plt.show()
```
